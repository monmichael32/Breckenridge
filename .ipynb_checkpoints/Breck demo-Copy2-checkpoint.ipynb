```python
def get_dates():
  #list_1 = []
  #list_2 = []
  #list_3 = []
  Calendar = driver.find_elements_by_class_name("selectable")
  Date = 0
  #print("Calendar", Calendar)

  for myDate in Calendar: 
    try:
        Calendar = driver.find_elements_by_class_name("selectable")
        D = (myDate.find_element_by_tag_name("a"))
        Date = int(D.text)
        list_2.extend([Date])
        driver.implicitly_wait(3)
    except:
        pass
  return list_2

def get_date_webelements(): 
  Calendar = driver.find_elements_by_class_name("selectable")
  #print("Calendar", Calendar)
  #list_1 = []
  for myDate in Calendar:     
    #print ("myDate: ", myDate)
    try:
        Calendar = driver.find_elements_by_class_name("selectable")
        #list_1 = [(driver.find_elements_by_class_name("selectable").myDate.find_elements_by_tag_name("a"))]
        D = (myDate.find_element_by_tag_name("a"))
        list_1.extend([D])
        driver.implicitly_wait(3)
    except:
        pass
  return list_1

def get_date_webelements_deselect():
    WD = driver.find_elements_by_class_name("ui-datepicker-calendar")
    list_4 = []
    for myDate in WD:
      try:
         WD = driver.find_elements_by_class_name("ui-datepicker-calendar")
         wd = (myDate.find_element_by_tag_name("a"))
         list_4.extend([wd])
      except:
          pass
    #print("list in get date:",list_4)
    return list_4
    #
    
    #
```
